```python
def sum_list(numbers):
    total = 0
    for x in numbers:
        total += x
    return total

sample_list = [5,8,9,10,11,20]
expected_output = sum_list(sample_list)

print(f"Sample List: {sample_list}")
print(f"Expected Output: {expected_output}")
```

```python
my_list = [10, 20, 30, 40, 50]
print(f"Original list: {my_list}")

# Get the number to be inserted from the user
try:
    number_to_insert = int(input("Enter the number to insert: "))
except ValueError:
    print("Invalid input. Please enter an integer for the number.")
    number_to_insert = None

# Get the position from the user
try:
    position_to_insert = int(input("Enter the position (0-indexed) to insert the number: "))
except ValueError:
    print("Invalid input. Please enter an integer for the position.")
    position_to_insert = None

# Insert the number if inputs are valid
if number_to_insert is not None and position_to_insert is not None:
    # Ensure position is within valid range
    if 0 <= position_to_insert <= len(my_list):
        my_list.insert(position_to_insert, number_to_insert)
        print(f"List after insertion: {my_list}")
    else:
        print(f"Invalid position. Please enter a position between 0 and {len(my_list)}.")
```

```python
sample_dict = { "name": "Kelly", "age":25, "salary": 8000, "city": "New york"}
print(f"Original dictionary: {sample_dict}")

# Get the value associated with the 'city' key
city_value = sample_dict.pop('city')

# Add a new key 'location' with the retrieved value
sample_dict['location'] = city_value

print(f"Dictionary after renaming 'city' to 'location': {sample_dict}")
```
